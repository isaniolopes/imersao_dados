```python
#seaborn
import seaborn as sns

sns.countplot(x = 'composto', data=dados)
```

```python
#5 drogas com maior frequencia
compostos = dados['composto'].value_counts()[0:5]
compostos
```

```python
#pegar somente os nomes dos compostos
cod_compostos = dados['composto'].value_counts().index[0:5]
cod_compostos
```

```python
#selecionar os dados dos 5 compostos mais frequentes
dados.query('composto in @cod_compostos')
```

```python
#grafico dos 5 compostos mais frequentes
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
#grafico dos 5 compostos mais frequentes, ordenado pela frequencia
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
#melhorar o gráfico, alterar tamanho da fonte
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 Compostos', fontsize=15)
ax.set_xlabel('Compostos', fontsize=15)
ax.set_ylabel('Frequência', fontsize=15)
plt.show()
```

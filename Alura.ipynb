```python
#gerando novos indices somente com o moa. Eu separo cada indice por '_' e depois escolho a ultima palavra [-1]
novos_indices = [i.split('_')[-1] for i in contagem_moa.index ]
novos_indices[:10]
```

```python
#montando novo df com as colunas aÃ§Ã£o (e seus valores sendo os novos indices) e contagem (valores presentes na variÃ¡vel contagem_moa)
novo_contagem_moa = pd.DataFrame(data={'aÃ§Ã£o':novos_indices,'contagem':contagem_moa.values})
novo_contagem_moa
```

```python
#agrupo por aÃ§Ã£o, somo todos os valores para as respectivaÃ§Ãµes e depois organizo por ordem de contagem e seleciono o top10
top10_aÃ§Ãµes = novo_contagem_moa.groupby(by=['aÃ§Ã£o']).sum()['contagem'].sort_values(ascending=False)[:10]
plt.figure(figsize=(15,7))
top10_aÃ§Ãµes.plot.bar()
```

```python
#criar agrupamento por composto e retornar quantidade de valores unicos de ativo_moa
grupo_ativo_moa = dados_combinados.groupby(by='composto')['ativo_moa'].nunique()
grupo_ativo_moa
```

```python
#conferir se algum composto por acaso ativa e nÃ£o ativa genes dependendo da condiÃ§Ã£o
count = 0
for i in range(len(grupo_ativo_moa)):
  if grupo_ativo_moa[i] == 1:
    pass
  else:
    count += 1
    print(grupo_ativo_moa[i].index)

if count == 0:
  print('Todos os compostos apresentam somentem um tipo de ativaÃ§Ã£o de moa')
else:
  pass
```

```python
#fazer o mesmo para o n_moa
grupo_n_moa = dados_combinados.groupby(by='composto')['n_moa'].nunique()

count = 0
for i in range(len(grupo_n_moa)):
  if grupo_n_moa[i] == 1:
    pass
  else:
    count += 1
    print(grupo_n_moa[i].index)

if count == 0:
  print('Todos os compostos sempre ativam (ou nÃ£o ativam) a mesma quantidade de mecanismos')
else:
  pass
```
